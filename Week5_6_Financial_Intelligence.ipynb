# Expense Intelligence

Analyze user spending patterns.

```python
# Sample expense data

import pandas as pd
import matplotlib.pyplot as plt

expense_data = {
    "Category": [
        "Food",
        "Shopping",
        "Travel",
        "Entertainment",
        "Transfer"
    ],

    "Amount": [
        2500,
        4000,
        1500,
        2000,
        3000
    ]
}

df = pd.DataFrame(expense_data)

print(df)
```

# Spending Analysis

Analyze total and average expenses.

```python
# Spending analysis

total_expense = df["Amount"].sum()
average_expense = df["Amount"].mean()

print("Total Expense: ₹", total_expense)
print("Average Expense: ₹", average_expense)
```

# Smart Expense Insights

Generate spending insights from expense data.

```python
# Expense insights

highest_expense = df.loc[df["Amount"].idxmax()]
lowest_expense = df.loc[df["Amount"].idxmin()]

print("Highest Spending Category:")
print(highest_expense["Category"], "-", "₹", highest_expense["Amount"])

print("\nLowest Spending Category:")
print(lowest_expense["Category"], "-", "₹", lowest_expense["Amount"])
```

# Personalized Financial Advice

Generate financial recommendations from spending patterns.

```python
# Personalized recommendation system

def personalized_advice(df):

    highest_expense = df.loc[df["Amount"].idxmax()]

    category = highest_expense["Category"]
    amount = highest_expense["Amount"]

    print("Personalized Financial Advice:\n")

    print(
        f"You are spending more on {category} "
        f"(₹{amount}). Try reducing unnecessary expenses "
        f"and improve monthly savings."
    )

# Run recommendation
personalized_advice(df)
```

# Financial Health Score

Calculate financial health based on spending patterns.

```python
# Financial health scoring

def financial_health_score(total_expense):

    if total_expense <= 5000:
        score = 90

    elif total_expense <= 10000:
        score = 75

    elif total_expense <= 20000:
        score = 60

    else:
        score = 40

    return score

# Calculate score
score = financial_health_score(total_expense)

print("Financial Health Score:", score, "/100")
```

# Expense Dashboard

Visualize expense distribution and spending trends.

```python
# Expense dashboard charts

# Category expense chart
df.plot(
    x="Category",
    y="Amount",
    kind="bar",
    figsize=(7,4),
    title="Expense Distribution"
)

plt.ylabel("Amount")
plt.show()
```

# Budget Status

Check whether spending is under control.

```python
# Budget status analysis

monthly_budget = 15000

if total_expense <= monthly_budget:

    print("Budget Status: Under Control")

else:

    print("Budget Status: Budget Limit Exceeded")
```

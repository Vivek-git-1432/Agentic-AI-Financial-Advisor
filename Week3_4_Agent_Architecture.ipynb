# Financial Recommendation Workflow

Create a basic financial advisory workflow using LangChain.

```python
# Import LangChain modules

from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

print("LangChain modules loaded successfully!")
```

# Financial Advice Prompt

Create prompt template for financial recommendations.

```python
# Create financial advice prompt

financial_prompt = PromptTemplate(
    input_variables=["amount", "category"],

    template="""
    You are a financial advisor.

    Expense Amount: в‚№{amount}
    Expense Category: {category}

    Give short financial advice.
    """
)

print("Financial prompt created successfully!")
```

# Generate Financial Advice

Generate recommendations from expense details.

```python
# Format financial advice prompt

formatted_prompt = financial_prompt.format(
    amount=500,
    category="Transfer"
)

print(formatted_prompt)
```

# Financial Recommendation Function

Create reusable financial recommendation workflow.

```python
# Financial recommendation function

def generate_financial_advice(amount, category):

    advice = f"""
    Expense Amount: в‚№{amount}
    Category: {category}

    Recommendation:
    Track your spending regularly and maintain savings.
    """

    return advice

print("Financial recommendation function created!")
```

# Test Financial Workflow

Test the recommendation workflow with sample expense data.

```python
# Test recommendation function

result = generate_financial_advice(
    amount=500,
    category="Transfer"
)

print(result)
```

# Financial Tool

Create a LangChain financial analysis tool.

```python
# Create financial analysis tool

from langchain.tools import Tool

financial_tool = Tool(
    name="Financial Advisor",

    func=lambda text: generate_financial_advice(
        amount=500,
        category="Transfer"
    ),

    description="Provides financial recommendations."
)

print("Financial tool created successfully!")
```

# Test Financial Tool

Run the financial analysis tool.

```python
# Execute financial tool

tool_output = financial_tool.run("Analyze my expense")

print(tool_output)
```

# Savings Tool

Create savings recommendation tool.

```python
# Savings recommendation function

def savings_advice():

    return """
    Savings Recommendation:
    Save at least 20% of monthly income.
    Avoid unnecessary spending.
    """

print("Savings function created!")
```

# Savings Recommendation Tool

Convert savings workflow into LangChain tool.

```python
# Create savings tool

savings_tool = Tool(
    name="Savings Advisor",

    func=lambda text: savings_advice(),

    description="Provides savings recommendations."
)

print("Savings tool created successfully!")
```

# Test Savings Tool

Execute savings recommendation tool.

```python
# Run savings tool

savings_output = savings_tool.run("Give savings advice")

print(savings_output)
```

# Budget Planning Tool

Create budget recommendation workflow.

```python
# Budget recommendation function

def budget_advice():

    return """
    Budget Recommendation:
    Follow a monthly budget plan.
    Track essential and non-essential expenses.
    """

print("Budget function created!")
```

# Budget Recommendation Tool

Convert budget workflow into LangChain tool.

```python
# Create budget tool

budget_tool = Tool(
    name="Budget Planner",

    func=lambda text: budget_advice(),

    description="Provides budgeting recommendations."
)

print("Budget tool created successfully!")
```

# Test Budget Tool

Execute budget recommendation tool.

```python
# Run budget tool

budget_output = budget_tool.run("Give budget advice")

print(budget_output)
```

# Multiple Financial Tools

Combine financial tools for AI advisory workflow.

```python
# Combine tools

financial_tools = [
    financial_tool,
    savings_tool,
    budget_tool
]

print("All financial tools combined successfully!")
```

# Financial Agent Workflow

Run multiple financial advisory tools.

```python
# Execute all tools

for tool in financial_tools:

    print(f"\nTool: {tool.name}\n")

    output = tool.run("Financial analysis")

    print(output)
```

# ReAct Agent Logic

Create simple reasoning workflow for financial recommendations.

```python
# Simple ReAct-style reasoning

def react_financial_agent(user_query):

    if "save" in user_query.lower():
        return savings_tool.run(user_query)

    elif "budget" in user_query.lower():
        return budget_tool.run(user_query)

    else:
        return financial_tool.run(user_query)

print("ReAct financial agent created successfully!")
```

# Test ReAct Financial Agent

Test agent reasoning with different financial queries.

```python
# Test agent queries

query_1 = "How can I save money?"
query_2 = "Give me budget advice"
query_3 = "Analyze my expenses"

print(react_financial_agent(query_1))
print("\n-------------------\n")

print(react_financial_agent(query_2))
print("\n-------------------\n")

print(react_financial_agent(query_3))
```

# Financial Knowledge Retrieval

Create financial knowledge base and retrieval workflow.

```python
# Financial knowledge retrieval

financial_knowledge = [
    "Saving money regularly improves financial stability.",

    "A monthly budget helps control unnecessary expenses.",

    "Emergency funds are important for financial safety.",

    "Tracking expenses helps improve spending habits.",

    "Investing regularly helps achieve long-term financial goals."
]

# Retrieval function
def retrieve_financial_knowledge(query):

    results = []

    for text in financial_knowledge:

        if any(word.lower() in text.lower() for word in query.split()):
            results.append(text)

    return results

# Test retrieval
retrieved_data = retrieve_financial_knowledge(
    "save money"
)

print("Retrieved Knowledge:\n")

for item in retrieved_data:
    print("-", item)
```

# AI Financial Assistant

Combine retrieval and financial recommendation workflow.

```python
# AI financial assistant workflow

def ai_financial_assistant(query):

    # Retrieve financial knowledge
    retrieved_data = retrieve_financial_knowledge(query)

    # Generate recommendation
    recommendation = react_financial_agent(query)

    # Final response
    print("Financial Recommendation:\n")
    print(recommendation)

    print("\nRetrieved Knowledge:\n")

    for item in retrieved_data:
        print("-", item)

# Test assistant
ai_financial_assistant("How can I save money?")
```

# OCR Text Extraction

This step extracts transaction text from the uploaded screenshot using Tesseract OCR.

```python
# Convert image for OCR

opencv_image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Extract text using OCR
extracted_text = pytesseract.image_to_string(opencv_image)

print("Extracted Text:\n")
print(extracted_text)
```

# Amount Detection

Detect transaction amount from OCR extracted text.

```python
# Extract only transaction amounts

amount_pattern = r'₹\s?(\d+)|\b(\d{2,5})\b'

matches = re.findall(amount_pattern, extracted_text)

amounts = []

for match in matches:
    for value in match:
        if value:
            amounts.append(int(value))

# Keep realistic payment amounts only
amounts = [amt for amt in amounts if amt < 100000]

detected_amount = max(amounts)

print("Detected Transaction Amount: ₹", detected_amount)
```

# Expense Category Detection

Automatically classify the expense category from transaction text.

```python
# Detect expense category

expense_categories = {
    "Food": ["swiggy", "zomato", "restaurant", "food"],
    "Shopping": ["amazon", "flipkart", "shopping", "store"],
    "Recharge": ["recharge", "mobile", "jio", "airtel"],
    "Travel": ["uber", "ola", "bus", "train"],
    "Entertainment": ["movie", "netflix", "spotify"],
    "Transfer": ["paid to", "upi", "bank transfer"]
}

detected_category = "Others"

lower_text = extracted_text.lower()

for category, keywords in expense_categories.items():
    for keyword in keywords:
        if keyword in lower_text:
            detected_category = category
            break

print("Detected Expense Category:", detected_category)
```

# Structured Expense Record

Store extracted expense information in structured format.

```python
# Create structured expense record

expense_record = {
    "Amount": detected_amount,
    "Category": detected_category
}

print(expense_record)
```

# Expense Database

Store expense records using SQLite database.

```python
# Import SQLite

import sqlite3

print("SQLite loaded successfully!")
```

# Create Expense Table

Create database table for storing expense records.

```python
# Create SQLite database and table

connection = sqlite3.connect("expenses.db")

cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS expenses (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    amount INTEGER,
    category TEXT
)
""")

connection.commit()

print("Expense table created successfully!")
```

# Save Expense Record

Insert extracted expense information into database.

```python
# Insert expense record into database

cursor.execute("""
INSERT INTO expenses (amount, category)
VALUES (?, ?)
""", (detected_amount, detected_category))

connection.commit()

print("Expense record saved successfully!")
```

# Expense Records Table

Display stored expense records in table format.

```python
import pandas as pd
```

```python
# Load expense records

df = pd.read_sql_query(
    "SELECT * FROM expenses",
    connection
)

df
```

# Expense Analytics

Analyze expenses based on categories.

```python
# Expense category summary

category_summary = df.groupby("category")["amount"].sum()

print(category_summary)
```

# Expense Visualization

Visualize expense distribution using charts.

```python
# Plot expense category chart

category_summary.plot(
    kind="bar",
    figsize=(6,4),
    title="Expense Distribution"
)

plt.xlabel("Category")
plt.ylabel("Amount")

plt.show()
```
